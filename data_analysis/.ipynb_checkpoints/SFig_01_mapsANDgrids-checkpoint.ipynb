## Load grids

### Ice/Ocean - 1deg

```python
# set a dataset to read in to get grid information
dir_in = '/glade/u/home/duvivier/masks/'
file_in = 'ocn_grid_gx1v7.nc'
ds = xr.open_mfdataset(dir_in+file_in, decode_times=False)

# grab lat and lon data
lat2d = ds.TLAT
lon2d = ds.TLONG
land = ds.KMT
```

```python
# First, find the index of the grid point nearest a specific lat/lon.   
abslat = np.abs(lat2d - lat_village)
abslon = np.abs(lon2d - lon_village)
c = np.maximum(abslon, abslat)

# find the smallest values
([xloc], [yloc]) = np.where(c == np.min(c))

# print index values - we know from before that these index values are correct and closest ocean points to UTQ
print(xloc,yloc)
```

```python
# save some data for plotting later
land_ocn_lr = land

lon2d_ocn_lr = lon2d
lat2d_ocn_lr = lat2d

xloc_ocn_lr = xloc
yloc_ocn_lr = yloc
```

### Ice/Ocean - 0.1deg 

```python
# set a dataset to read in to get grid information
dir_in = '/glade/u/home/duvivier/masks/'
file_in = 'POP_tx0.1v3_grid.nc'
ds_grid = xr.open_mfdataset(dir_in+file_in, decode_times=False)

# rename coordinates to those used in ice files
ds_grid = ds_grid.rename({'nlat': 'nj', 'nlon': 'ni'})
# clip global data to just Alaska
ds_grid_sub = ds_grid.isel(nj=slice(1700,2250)).isel(ni=slice(2700,3300))
```

```python
# First, find the index of the grid point nearest a specific lat/lon.   
abslat = np.abs(lat2d - lat_village)
abslon = np.abs(lon2d - lon_village)
c = np.maximum(abslon, abslat)

# find the smallest values
([xloc], [yloc]) = np.where(c == np.min(c))

# print index values
print(xloc,yloc)
```

```python
# save some data for plotting later
land_ocn_hr = land

lon2d_ocn_hr = lon2d
lat2d_ocn_hr = lat2d

xloc_ocn_hr = xloc
yloc_ocn_hr = yloc
```

### Atm - 1deg

```python
# set a dataset to read in to get grid information
dir_in = '/glade/u/home/duvivier/masks/'
file_in = 'b.e21.BSSP370.f09_g17.rufmod.001.cam.h0.2015-01.nc'
ds = xr.open_mfdataset(dir_in+file_in, decode_times=False)
# keep only one time
ds = ds.isel(time=0)
```

```python
# grab lat and lon data
lat1d = ds.lat
lon1d = ds.lon
land = ds.LANDFRAC
wind = ds.U10
```

```python
# First, find the index of the grid point nearest a specific lat/lon.   
abslat = np.abs(lat1d - lat_village)
abslon = np.abs(lon1d - lon_village)

# find the smallest values
xloc = np.where(abslat == np.min(abslat))
yloc = np.where(abslon == np.min(abslon))

# print index values
print(xloc[0],yloc[0])
```

```python
#create a meshgrid (2D fields of lats and lons)
lon2d,lat2d=np.meshgrid(lon1d,lat1d)

# convert the numpy array to an xarray
lon2d = xr.DataArray(lon2d,dims=('lat','lon'))
lat2d = xr.DataArray(lat2d,dims=('lat','lon'))

# assign coordinates
lon2d['lat'] = land.lat
lon2d['lon'] = land.lon
lat2d['lat'] = land.lat
lat2d['lon'] = land.lon
```

```python
# save some data for plotting later
wind_atm_lr = wind

lon2d_atm_lr = lon2d
lat2d_atm_lr = lat2d

xloc_atm_lr = int(xloc[0])+1
yloc_atm_lr = int(yloc[0])
```

```python
#create a meshgrid (2D fields of lats and lons)
lon2d,lat2d=np.meshgrid(lon1d,lat1d)

# convert the numpy array to an xarray
lon2d = xr.DataArray(lon2d,dims=('lat','lon'))
lat2d = xr.DataArray(lat2d,dims=('lat','lon'))

# assign coordinates
lon2d['lat'] = ds_grid.lat
lon2d['lon'] = ds_grid.lon
lat2d['lat'] = ds_grid.lat
lat2d['lon'] = ds_grid.lon
```

```python
# save some data for plotting later
temp_atm_hr = temp - 273.15

lon2d_atm_hr = lon2d
lat2d_atm_hr = lat2d

xloc_atm_hr = int(xloc[0])+1
yloc_atm_hr = int(yloc[0])
```

### 3deg Waves

```python
# set a dataset to read in to get grid information - using grid file that Frank Bryan made and provided
dir_in = '/glade/u/home/duvivier/masks/'
file_in = 'ww3.4x3.grid.nc'
ds = xr.open_dataset(dir_in+file_in, decode_times=False)
```

```python
# grab lat and lon data
lat1d = ds.lat
lon1d = ds.lon
mask = ds.mask

#create a meshgrid (2D fields of lats and lons)
lon2d,lat2d=np.meshgrid(lon1d,lat1d)

# create an array to plot
ww3_field = ((lat2d * lon2d)/100)
```

```python
# First, find the index of the grid point nearest a specific lat/lon.   
abslat = np.abs(lat1d - lat_village)
abslon = np.abs(lon1d - lon_village)
c = np.maximum(abslon, abslat)

# find the smallest values
([xloc], [yloc]) = np.where(c == np.min(c))

# print index values
print(xloc,yloc)
```

```python
# save some data for plotting later
field_ww3 = ww3_field

lon2d_ww3 = lon2d
lat2d_ww3 = lat2d

xloc_ww3 = xloc
yloc_ww3 = yloc
```
